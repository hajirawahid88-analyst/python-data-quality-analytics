```python
# --- Count invalid CNICs per sector ---
invalid_cnic_summary = (
    df[df["CNIC Validation Status"] == "Invalid"]
    .groupby("Area")
    .size()
    .reset_index(name="Invalid_CNIC_Count")
    .sort_values("Invalid_CNIC_Count", )
)

# --- Count invalid Mobile Numbers per sector ---
invalid_mobile_summary = (
    df[df["Mobile Validation Status"] == "Invalid"]
    .groupby("Area")
    .size()
    .reset_index(name="Invalid_Mobile_Count")
    .sort_values("Invalid_Mobile_Count", )
)

# --- Display or export the results ---
print("Invalid CNIC Count per Sector:")
print(invalid_cnic_summary)

print("\nInvalid Mobile Count per Sector:")
print(invalid_mobile_summary)
```
