```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data for heatmap
heatmap_data = Area_summary.set_index('Area')[['Percentage_Invalid_CNIC', 'Percentage_Invalid_Mobile']]

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='YlOrRd', fmt=".2f", linewidths=.5)
plt.title('Heatmap of of voilations by Area', fontsize=16)
plt.xlabel('Validation Type', fontsize=12)
plt.ylabel('Area', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
