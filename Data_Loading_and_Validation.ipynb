```python
import pandas as pd

# Replace 'your_excel_file.xlsx' with the actual path to your file in Google Drive
excel_file_path = '//content/drive/MyDrive/Operational Data.xlsx'
#excel_file_path = '/content/october_comprehensive.xlsx'

try:
    df = pd.read_excel(excel_file_path, dtype={"cnic": str, "mobile number": str})
    print("Excel file read successfully!")
    #print(df)
except FileNotFoundError:
    print(f"Error: The file '{excel_file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
def validate_cnic(cnic):
    cnic = str(cnic or "").replace("-", "").strip()
    if (cnic == "" or
        len(cnic) != 13 or
        not cnic.isdigit() or
        cnic.startswith("000") or
        cnic.startswith("123")):
        return "Invalid"
    else:
        return "Valid"
```

```python
def validate_mobile(mobile):
    mobile = str(mobile or "").replace("-", "").strip()
    if (mobile == "" or
        len(mobile) != 11 or
        not mobile.isdigit() or
        mobile.startswith("000") or
        mobile.startswith("123")):
        return "Invalid"
    else:
        return "Valid"
```

```python
df['CNIC Validation Status'] = df.apply(lambda row: validate_cnic(row['cnic']), axis=1)
print("CNIC validation applied")
df['Mobile Validation Status'] = df.apply(lambda row: validate_mobile(row['mobile number']), axis=1)
print("Mobile validation applied")
```
